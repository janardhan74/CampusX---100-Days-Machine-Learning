```python
import pandas as pd
from sklearn.datasets import make_classification
```

```python
x,y = make_classification(n_samples=500,n_features=5,n_informative=3,n_redundant=2,n_clusters_per_class=1)
```

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import BaggingClassifier
from sklearn.ensemble import RandomForestClassifier
```

```python
bc = BaggingClassifier(estimator=DecisionTreeClassifier(),max_features=2)
bc.fit(x,y)
```

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(16,8))
from sklearn.tree import plot_tree
plot_tree(bc.estimators_[0])
```

```python
rfc = RandomForestClassifier()
rfc.fit(x,y)
```

```python
rfc.estimators_
```

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(16,8))
from sklearn.tree import plot_tree
plot_tree(rfc.estimators_[0])
```
