```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('wine_data.csv',header=None,usecols=[0,1,2])
df
```

```python
df.columns=['Class label','Alchol','Malic Acid']
df
```

```python
sns.kdeplot(df['Alchol'])
```

```python
sns.kdeplot(df['Malic Acid'])
```

```python
sns.scatterplot(data=df,x='Alchol',y='Malic Acid',hue='Class label',palette={1:"red",2:"blue",3:"green"})
```

```python
from sklearn.model_selection import train_test_split
x = df.drop(['Class label'],axis=1)
y = df['Class label']
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.3,random_state=0)
x_train.shape , x_test.shape
```

```python
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
scaler.fit(x_train)

x_train_scaled = scaler.transform(x_train)
x_test_scaled = scaler.transform(x_test)
x_train_scaled
```

```python
x_train_scaled = pd.DataFrame(x_train_scaled,columns=x_train.columns)
x_test_scaled = pd.DataFrame(x_test_scaled,columns=x_train.columns)
x_train_scaled
```

### Effect of Scaling

```python
fig,(ax1,ax2) = plt.subplots(nrows=1,ncols=2,figsize=(12,5))

sns.scatterplot(x=x_train['Alchol'],y=x_train_scaled['Malic Acid'],ax=ax1,hue=y_train)
ax1.set_title('Before Scaling')

sns.scatterplot(x=x_train_scaled['Alchol'],y=x_train_scaled['Malic Acid'],ax=ax2,hue=y_train)
ax2.set_title('After Scaling')
```

```python
fig,(ax1 , ax2) = plt.subplots(ncols=2,nrows=1,figsize=(12,5))

sns.kdeplot(data=x_train['Alchol'],ax=ax1)
sns.kdeplot(data=x_train['Malic Acid'],ax=ax1)
ax1.set_title('Before Scaling')

sns.kdeplot(data=x_train_scaled['Alchol'],ax=ax2)
sns.kdeplot(data=x_train_scaled['Malic Acid'],ax=ax2)
ax2.set_title('After Scaling')

# after scalling -> the two features lies in same scale
```

```python
fig,(ax1 , ax2) = plt.subplots(ncols=2,nrows=1,figsize=(12,5))

sns.kdeplot(data=x_train['Alchol'],ax=ax1)
ax1.set_title('Before Scaling')

sns.kdeplot(data=x_train_scaled['Alchol'],ax=ax2)
ax2.set_title('After Scaling')
```

```python
fig,(ax1 , ax2) = plt.subplots(ncols=2,nrows=1,figsize=(12,5))

sns.kdeplot(data=x_train['Malic Acid'],ax=ax1)
ax1.set_title('Before Scaling')

sns.kdeplot(data=x_train_scaled['Malic Acid'],ax=ax2)
ax2.set_title('After Scaling')
```
