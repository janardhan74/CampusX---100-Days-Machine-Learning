```python
import pandas as pd
```

```python
df = pd.read_csv('mnist_train.csv').iloc[:1000,:]
df.head()
```

```python
df.shape
```

```python
import seaborn as sns
x = df.iloc[:,1:].values
y = df.iloc[:,0].values
sns.heatmap(x[0].reshape(28,28))
```

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
rfc = RandomForestClassifier()
rfc.fit(x,y)
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(y,rfc.predict(x))
```

```python
sns.heatmap(rfc.feature_importances_.reshape(28,28))
```
