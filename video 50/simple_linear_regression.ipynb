```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
```

```python
df = pd.read_csv('placements.csv')
df.head()
```

```python
plt.scatter(x=df['cgpa'],y=df['package'])
plt.xlabel("cgpa")
plt.ylabel("package")
# the data looke like linear data
```

```python
x = df[['cgpa']]
y = df[['package']]
x_train,x_test,y_train,y_test = train_test_split(x,y,random_state=42)
x_train.shape,x_test.shape
```

```python
lin_r = LinearRegression()
lin_r.fit(x_train,y_train)
```

```python
y_pred = lin_r.predict(x_test)
# accuracy_score(y_test,y_pred) 
# -> getting - ValueError: continuous is not supported on above line : TODO
```

```python
plt.scatter(x=df['cgpa'],y=df['package'])

plt.plot(df['cgpa'],lin_r.predict(df[['cgpa']]),color='red')

# here line represents the best fit line of Linear Regression for the given data
```

```python
m = lin_r.coef_
b = lin_r.intercept_
```
