```python
import pandas as pd
```

```python
boston = pd.read_csv('HousingData.csv')
boston.dropna(inplace=True)
boston
```

```python
from sklearn.tree import DecisionTreeRegressor
from sklearn.svm import SVR
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import cross_val_score
```

```python
lr = LinearRegression()
dt = DecisionTreeRegressor()
svr = SVR()

x = boston.iloc[:,:-1].values
y = boston.iloc[:,-1].values
estimators = [('lr',lr),('dt',dt),('svr',svr)]

for name,obj in estimators:
    r2 = cross_val_score(obj,x,y,cv=10,scoring='r2').mean()
    print(name,r2)
```

```python
from sklearn.ensemble import VotingRegressor
estimators = [('lr',lr),('dt',dt),('svr',svr)]
vr = VotingRegressor(estimators=estimators)
r2 = cross_val_score(vr,x,y,cv=10,scoring='r2').mean()
r2
```

```python
import numpy as np

for i in range(1,4):
    for j in range(1,4):
        for k in range(1,4):
            vr = VotingRegressor(estimators=estimators,weights=[i,j,k])
            r2 = cross_val_score(vr,x,y,cv=10,scoring='r2').mean()
            print(i,j,k,np.round(r2,2))
```

```python
dt1 = DecisionTreeRegressor(max_depth=1)
dt2 = DecisionTreeRegressor(max_depth=2)
dt3 = DecisionTreeRegressor(max_depth=3)
dt4 = DecisionTreeRegressor(max_depth=4)
dt5 = DecisionTreeRegressor(max_depth=5)

estimators = [('dt1',dt1),('dt2',dt2),('dt3',dt3),('dt4',dt4),('dt5',dt5)]
for name,obj in estimators:
    r2 = cross_val_score(obj,x,y,cv=10,scoring='r2').mean()
    print(name,r2)
```

```python
vr = VotingRegressor(estimators = estimators)
cross_val_score(vr,x,y,cv=10,scoring='r2').mean()
```
