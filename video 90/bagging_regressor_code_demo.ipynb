```python
import pandas as pd
from sklearn.model_selection import train_test_split,GridSearchCV
from sklearn.ensemble import BaggingRegressor
from sklearn.linear_model import LinearRegression
from sklearn.neighbors import KNeighborsRegressor
from sklearn.tree import DecisionTreeRegressor
from sklearn.metrics import r2_score
```

```python
df = pd.read_csv('HousingData.csv')
df.dropna(inplace=True)
df.head()
```

```python
x = df.iloc[:,:-1]
y = df.iloc[:,-1]
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
lr = LinearRegression()
dt = DecisionTreeRegressor()
knn = KNeighborsRegressor()
```

```python
lr.fit(x_train,y_train)
dt.fit(x_train,y_train)
knn.fit(x_train,y_train)
```

```python
print("Accuracy lr",r2_score(y_test,lr.predict(x_test)))
print("Accuracy dt",r2_score(y_test,dt.predict(x_test)))
print("Accuracy knn",r2_score(y_test,knn.predict(x_test)))
```

```python
# Bagging
```

```python
bag = BaggingRegressor()
bag.fit(x_train,y_train)
```

```python
print("Accuracy dt",r2_score(y_test,bag.predict(x_test)))
```

```python
# grid search cv
param_grid = {
    'estimator':[None,KNeighborsRegressor(),LinearRegression()],
    'n_estimators':[20,50,100],
    'max_samples':[0.5,1.0],
    'max_features':[0.5,1.0],
    'bootstrap':[True,False],
    'bootstrap_features':[True,False]
}
gs = GridSearchCV(BaggingRegressor(n_jobs=-1),param_grid=param_grid,cv=3,n_jobs=-1,verbose=1)
gs.fit(x_train,y_train)
```
